## ¿Qué es una estructura?

Una estructura permite agrupar variables relacionadas en un solo tipo de dato.

Esto es muy común en sistemas embebidos para representar dispositivos como:

- sensores
- motores
- LEDs
- periféricos

## Ejemplo básico de struct en C

````python
codigo = '''
#include <stdio.h>

struct Sensor {
    int id;
    float temperatura;
};

int main() {

    struct Sensor s1;

    s1.id = 1;
    s1.temperatura = 25.5;

    printf("Sensor %d Temp %.2f", s1.id, s1.temperatura);

}
'''

print(codigo)
````

## Uso de typedef struct

`typedef` permite crear un alias del tipo de dato para simplificar el código.

````python
codigo = '''
typedef struct {

    int pin;
    int estado;

} LED;

LED led1;
'''

print(codigo)
````

## Arreglo de estructuras

También podemos crear arreglos de estructuras.

````python
codigo = '''
LED leds[3];

leds[0].pin = 1;
leds[1].pin = 2;
leds[2].pin = 3;
'''

print(codigo)
````
